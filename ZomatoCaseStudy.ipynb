```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (update the file path accordingly)
df = pd.read_csv("zomato.csv")

# Convert ratings to numeric values
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False)  
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')  

# Group by restaurant name and get the average rating
restaurant_ratings = df.groupby('name')['rate'].mean().dropna()

# Sort by rating and select top 20 restaurants
top_20_restaurants = restaurant_ratings.sort_values(ascending=False).head(20)

# Plot horizontal bar graph (Fixed Palette Issue)
plt.figure(figsize=(12, 8))
sns.barplot(y=top_20_restaurants.index, x=top_20_restaurants.values, hue=top_20_restaurants.index, dodge=False, legend=False, palette="viridis")

# Labels and title
plt.ylabel("Restaurant Name", fontsize=12, fontweight='bold')
plt.xlabel("Average Rating", fontsize=12, fontweight='bold')
plt.title("Top 20 Restaurants vs Ratings", fontsize=14, fontweight='bold')

# Show plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["votes"], y=df["rate"], alpha=0.6, color="purple")

# Labels & Title
plt.xlabel("Number of Votes", fontsize=12, fontweight="bold")
plt.ylabel("Ratings", fontsize=12, fontweight="bold")
plt.title("Votes vs. Ratings", fontsize=14, fontweight="bold")

plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x="online_order", y="rate", data=df, hue="online_order", palette="coolwarm", legend=False)

# Labels & Title
plt.xlabel("Online Order Available", fontsize=12, fontweight="bold")
plt.ylabel("Ratings", fontsize=12, fontweight="bold")
plt.title("Online Orders vs. Ratings", fontsize=14, fontweight="bold")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by a categorical column (e.g., 'online_order') and count values
data = df["online_order"].value_counts()

# Define aesthetic color palette
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0", "#ffb3e6"]

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, colors=colors, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}, textprops={'fontsize': 12})

# Title
plt.title("Distribution of Online Orders", fontsize=14, fontweight="bold")

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Replace 'zomato.csv' with your dataset file)
df = pd.read_csv("zomato.csv")

# Convert ratings to numeric values (if necessary)
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False)  # Remove '/5' from ratings
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')  # Convert to numeric

# Select relevant columns
df_filtered = df[['name', 'rate', 'votes', 'approx_cost(for two people)']].dropna()

# Get top 10 restaurants based on rating
top_restaurants = df_filtered.groupby('name').mean().sort_values(by='rate', ascending=False).head(10)

# Set professional style
sns.set_style("whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Bar Plot 1 - Restaurant Ratings
sns.barplot(x=top_restaurants['rate'], y=top_restaurants.index, ax=axes[0], hue=top_restaurants.index, palette="magma", legend=False)
axes[0].set_title("Top 10 Restaurants - Ratings", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Average Rating", fontsize=12)
axes[0].set_ylabel("Restaurant Name", fontsize=12)

# Bar Plot 2 - Number of Votes
sns.barplot(x=top_restaurants['votes'], y=top_restaurants.index, ax=axes[1], hue=top_restaurants.index, palette="coolwarm", legend=False)
axes[1].set_title("Top 10 Restaurants - Votes", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Number of Votes", fontsize=12)
axes[1].set_ylabel("")

# Bar Plot 3 - Approx Cost for Two
sns.barplot(x=top_restaurants['approx_cost(for two people)'], y=top_restaurants.index, ax=axes[2], hue=top_restaurants.index, palette="viridis", legend=False)
axes[2].set_title("Top 10 Restaurants - Approx Cost", fontsize=14, fontweight='bold')
axes[2].set_xlabel("Cost for Two (â‚¹)", fontsize=12)
axes[2].set_ylabel("")

# Adjust layout
plt.tight_layout()
plt.show()
```
